```python
print(df['HS'].value_counts())
print(len(df['HS'].value_counts()))
```

```python
df['HS']=df['HS'].apply(lambda x:x[:len(x)-1] if x[-1]=="*" else x)
print(df['HS'].value_counts())
print(len(df['HS'].value_counts()))
print(len(df['HS']))
df['HS']=df['HS'].astype('int')
```

```python
print(df['Avg'].value_counts())
for i in df['Avg']:
  print(i)
# df['Avg']=df['Avg'].astype('float')
```

```python
df['Avg']=df['Avg'].replace('-',np.nan)
df['Avg']=df['Avg'].fillna(df['Avg'].median)
df['Avg']=df['Avg'].astype('float',copy=True)
```
